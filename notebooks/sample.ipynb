### Delete application inference profiles

> To housekeep the profiles.

```python
import time

response = client.list_inference_profiles(
    typeEquals='APPLICATION'
)
for inference_profile in response['inferenceProfileSummaries']:
    response = client.delete_inference_profile(
        inferenceProfileIdentifier=inference_profile['inferenceProfileArn']
    )
    time.sleep(0.5)
```
